```python
from src.services import AzureOpenaiService
from src.utils import Settings

sets = Settings()
azure_openai_service = AzureOpenaiService(sets)
llm = azure_openai_service.get_llm()
embeddings = azure_openai_service.get_embeddings()
```

```python
# llm call example
prompt = "What is the capital of France?"
response = llm.invoke(prompt)
print(response.content)
```

```python
# template and chain example
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant that translates the user text to {output_language}.",
        ),
        ("human", "{input}"),
    ]
)
chain = prompt | llm
res = chain.invoke(
    {
        "output_language": "Portuguese",
        "input": "I love programming.",
    }
)
print(res.content)
```

```python
# embeddings call example
text = "Paris is the capital of France."
embedding = embeddings.embed_query(text)
print(embedding)
```
